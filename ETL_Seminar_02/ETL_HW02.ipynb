### Задача 3

#### Напишите функцию, которая добавит колонку с разбиением возраста по группам: до 20, от 20 до 30, от 30 до 36 и старше 36. Посчитайте количество футболистов в каждой категории.

```python
df4['Age'].value_counts()
```

```python
# create a list of our conditions
conditions = [
    (df4['Age'] < 20),
    (df4['Age'] >= 20) & (df4['Age'] < 30),
    (df4['Age'] >= 30) & (df4['Age'] < 36),
    (df4['Age'] >= 36)
    ]

# create a list of the values we want to assign for each condition
values = ['Group 1', 'Group 2', 'Group 3', 'Group 4']

# create a new column and use np.select to assign values to it using our lists as arguments
df4['Age_Groups'] = np.select(conditions, values)

# display updated DataFrame
df4.head()
```

```python
df4['Age_Groups'].value_counts()
```
