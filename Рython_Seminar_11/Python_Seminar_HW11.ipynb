Свойства функции (sin(x)/2 + cos(x)/2) ** 2 * sin(x) и её график

```python
from sympy import *
from sympy.abc import x
from sympy.plotting import plot
```

```python
f = (sin(x)/2 + cos(x)/2) ** 2 * sin(x)
```

График функции

```python
plot(f, (x, -1, 1))
print(f)
```

```python
plot(f, (x, -5, 5))
print(f)
```

```python
plot(f, (x, -7.5, 7.5))
print(f)
```

5. Нули функции

```python
list_roots = solve(f, x) # но здесь нули функции указаны без учёта периода функции 2π. Только на одном периоде 2π.
list_roots = sorted(list_roots) # отсортирую по возрастанию
list_roots
```

```python
# в таких значениях мне проше воспринимать нули функции
for i in list_roots:
    print(round(float(i), 3))
```

Но и при использовании solveset тоже правильный ответ, но уже с учётом периода функции 2π.

Вообще в Интернете пишут следующее

SymPy has two solvers, solve (old, powerful, complicated, some might say messy) and solveset (new, clean, not yet as powerful but is supposed to be the future). I recommend working with solveset.

Поэтому в качестве ответа по нулям функции можно взять вывод команды с использованием solveset

```python
solveset(f, x, domain=S.Reals)
```

Для дальнейшего изучения свойств функции нам пригодится производная. 

```python
#найдём производную функции
diff_f = diff(f)
print(diff_f)
```

```python
# нарисуем график функции и ниже график производной (мне так лучше для понимания экстремумов и т.п.)
plot(f, (x, -5, 5))
print(f)
```

```python
plot(diff_f, (x, -5, 5))
print(diff_f)
```

```python
diff_roots = solve(diff_f, x, domain=S.Reals) #нахождение экстремумов
diff_roots = sorted(diff_roots) # отсортирую по возрастанию
diff_roots
```

```python
# в таких значениях мне проше воспринимать экстремумы функции
for i in diff_roots:
    print(round(float(i), 3))
```
